```python
subject = ['Eng','Math','Kor','Sci','Comp']
scores = [90,40,50,70,100]

# draw axis
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

# draw graph
ax1.bar(subject,scores)

# label, title, ... 
plt.xlabel("Subject")
plt.ylabel("Points")
plt.title("test result")

# show
plt.savefig('./barplot.png')   # save graph as image
plt.show()                     # print graph
```

```python
from datetime import datetime
import pandas as pd
import os

csv_path = "./data/AMZN2.csv"
data = pd.read_csv(csv_path ,index_col=0, parse_dates=True)
price = data['Close']  # Series

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
price.plot(ax=ax, style='black')
plt.ylim([1600,2200])
plt.xlim(['2019-05-01','2020-03-01'])

important_data = [(datetime(2019, 6, 3), "Low Price"),(datetime(2020, 2, 19), "Peak Price")]

# annotate : add letters or arrows
for d, label in important_data:
    ax.annotate(label, xy=(d, price.asof(d)+10), 
                xytext=(d,price.asof(d)+100),
                arrowprops=dict(facecolor='red'))

plt.grid()
ax.set_title('StockPrice')

plt.show()
```
